# Horizon to JSON

```python
import numpy as np
```

```python
raw = np.loadtxt('data/F3_horizon.txt')
```

```python
raw.shape
```

```python
nulls = np.amax(raw)
np.place(raw, raw==nulls, np.nan)
```

```python
ilinit, ilfin = np.amin(raw[:,0]), np.amax(raw[:,0])
xlinit, xlfin = np.amin(raw[:,1]), np.amax(raw[:,1])
nIL, nXL = ilfin-ilinit, xlfin-xlinit

print("inline range {} to {} ({} inlines)".format( ilinit, ilfin, ilfin - ilinit))
print("xline range {} to {} ({} xlines)".format(xlinit, xlfin, xlfin - xlinit))
```

```python
z = np.empty((nIL, nXL))
#amp = np.empty_like(z)
#sim = np.empty_like(z)

for i, row in enumerate(raw):
    il = row[0] - ilinit - 1
    xl = row[1] - xlinit - 1
    zi = row[2]
#    ai = row[3]
#    si = sim_raw[i][3]
    
    z[il, xl] = zi
#    amp[il, xl] = ai
#    sim[il, xl] = si
```
